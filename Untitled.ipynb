```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load world shapefile (you can download it from Natural Earth or use geopandas datasets)
world = gpd.read_file("ne_110m_admin_0_countries.zip")
world
```

```python
# Example statistics: number of mentions (replace this with your actual data)
mentions = {'Norway': 1, 'Sweden': 1, 'Finland': 1, 'Russia': 1}  # Example data
world['mentions'] = world['name'].map(mentions).fillna(1)  # Fill NaN with 1 to avoid scaling to zero

# Define a function to scale polygons
def scale_polygon(polygon, scale):
    if polygon.geom_type == 'Polygon':
        return gpd.GeoSeries([polygon]).scale(xfact=scale, yfact=scale, origin='centroid')[0]
    elif polygon.geom_type == 'MultiPolygon':
        return gpd.GeoSeries([polygon]).scale(xfact=scale, yfact=scale, origin='centroid')[0]

# Scale the polygons
world['geometry'] = world.apply(lambda x: scale_polygon(x['geometry'], x['mentions']), axis=1)

# Plotting
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
world.plot(ax=ax, color='lightblue', edgecolor='black')
plt.show()
```
