```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load world shapefile
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Example statistics: number of mentions
mentions = {'Norway': 10, 'Sweden': 5, 'Finland': 8, 'Russia': 20}
world['mentions'] = world['name'].map(mentions).fillna(1)  # Fill NaN with 1 to avoid missing data issues

# Plotting the world
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
world.boundary.plot(ax=ax, linewidth=1)

# Add scaled circles for mentions
for idx, row in world.iterrows():
    # Place circle at the centroid of the country
    centroid = row['geometry'].centroid
    ax.scatter(centroid.x, centroid.y, s=row['mentions'] * 100, alpha=0.5)  # Scale size of circle

plt.show()
```
